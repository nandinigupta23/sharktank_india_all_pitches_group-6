```python
def box_graph_bivar(colx,coly):
    plt.figure(figsize=(10,10))
    sns.boxplot(data=df,x=colx,y=coly,palette='cubehelix')
    plt.xticks(rotation=90)
    plt.show()
```

```python
import pandas as pd
df = pd.read_csv('ShartankIndiaAllPitches.csv')
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15, 3)
plt.subplot(1,3,1)
sns.barplot(x = df['Episode Number'],y = df['Aman'])
plt.subplot(1,3,2)
sns.barplot(x = df['Episode Number'],y = df['Pitch Number'])
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 3)
plt.subplot(1,3,1)
sns.barplot(x = df['Brand'],y = df['Episode Number'])
plt.subplot(1,3,2)
sns.barplot(x = df['Episode Number'],y = df['Season'])
plt.show()
```
